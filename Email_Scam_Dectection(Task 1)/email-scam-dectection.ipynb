### Adding columns for coverting catagorical values into numerical 

```python
encoder = LabelEncoder()
df['Catagory'] = encoder.fit_transform(df['Catagory'])
df
```

Here scam=1 & ham=0

## Creating Model

```python
#Splitting the data
X =  df["Message"]
y =df["Catagory"]
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size = 0.20, random_state = 0)
```

```python
coun = CountVectorizer()
X_train_count = coun.fit_transform(X_train.values)
X_train_count.toarray()
```

```python
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

## Accuracy of model

```python
# Traing Dataset
model.score(X_train_count,y_train)*100
```

```python
# Testing Dataset
X_test_count = coun.transform(X_test)
model.score(X_test_count,y_test)*100
```

```python
#Prediction
y=['Same. Wana plan a trip sometme then']
mail = coun.transform(y)
model.predict(mail)
```
